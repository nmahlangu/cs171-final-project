```python
def calc_time_array(schedule):
    time_array = []
    schedule_index = 0
    
    prev = None
    curr = schedule[schedule_index]

    for i in xrange(24 * 7):
                
        # start of the week
        if i < curr['day_order'] * 24 + curr['start_time'] and prev is None:
            time_array.append([curr['lat'], curr['lng'], 0])
            
        # during opening hours
        elif i >= (curr['day_order'] * 24 + curr['start_time']) and \
        i < (curr['day_order'] * 24 + curr['end_time']):
            time_array.append([curr['lat'], curr['lng'], 1]) 
            
        # during closing hours between opening days, 
        # UPDATE OF CURR AND PREV HAS ALREADY HAPPENED    
        # SO WE MUST NOW REFER TO PREV END TIME AND
        # CURR START TIME
        elif prev is not None and \
        i >= (prev['day_order'] * 24 + prev['end_time']) and \
        schedule_index < len(schedule) - 1 and \
        i < curr['day_order'] * 24 + curr['start_time']:
            time_array.append([prev['lat'], prev['lng'], 0])
            
        # closed for the rest of the week    
        elif schedule_index == len(schedule) - 1:
            time_array.append([curr['lat'], curr['lng'], 0])
            
        # during closing hours between opening days
        # UPDATES CURR AND PREV
        elif i >= (curr['day_order'] * 24 + curr['end_time']) and schedule_index < len(schedule) - 1:
            schedule_index += 1
            prev = curr
            curr = schedule[schedule_index]
            time_array.append([prev['lat'], prev['lng'], 0])
            
    return time_array
```

```python
truck_to_index = {}
index_to_truck = {}

test_id = 305709

time_table = []

index = 0
for truck_id, value in truck_data.iteritems():
    
    # index truck ids and which index they are in the time_table
    truck_to_index[truck_id] = index
    index_to_truck[index] = truck_id
    
    schedule = value['schedule']    
    schedule_168 = calc_time_array(schedule)
    
    time_table.append(schedule_168)
    index += 1
    
```
